```python


from sklearn.linear_model import Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor


models = {
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.01),
    "KNN": KNeighborsRegressor(n_neighbors=10),
    "SVM": SVR(kernel='poly'),  #'linear', 'poly', 'rbf', 'sigmoid', 'precomputed
    "DecisionTree": DecisionTreeRegressor(random_state=42)
}

results = {}

for name, model in models.items():
    # Use scaled data for KNN & SVM, unscaled for tree methods
    if name in ["KNN", "SVM", "Ridge", "Lasso"]:
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
    else:  # DecisionTree
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
    
    r2 = r2_score(y_test, y_pred)
    results[name] = r2

print("Model Performance (R² scores):")
for k, v in results.items():
    print(f"{k}: {v:.4f}")
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

selector = SelectKBest(score_func=f_regression, k=10)  # keep top 5 features
X_new = selector.fit_transform(x ,y)

selected_features = x.columns[selector.get_support()]
print("Selected Features:", selected_features)
```

```python
dt = DecisionTreeRegressor(max_depth=5, random_state=42)
dt.fit(x_train, y_train)
print("DecisionTree R²:", r2_score(y_test, dt.predict(x_test)))
```
